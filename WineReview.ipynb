## Data Cleaning

```python
# Check for null entries
wine_reviews.info()
```

We observe there are lot of null entries in the dataset. To make the process easier, we will replace the null entries with 'blanks'.

```python
# Fill all NA values with blank
wine_reviews.fillna("", inplace=True)
```

The main column for our prediction model would be the **description** column, which is a text entry. Therefore, it is essential that this column is clean, and can be used for the analysis. Thus, we will make a new column, **description_clean**, and remove any puctuations, digits, words of less than 3 characters. or english stop words, that might hinder our analysis of the text entered. Also, to ensure homogeneity, we will convert everything to lowercase.

```python
# Remove digits, punctuations, words of less than 3 characters
wine_reviews['description_clean'] = wine_reviews['description'].str.replace('\d+', '') # for digits
wine_reviews['description_clean'] = wine_reviews['description_clean'].str.replace(r'(\b\w{1,2}\b)', '') # for words
wine_reviews['description_clean'] = wine_reviews['description_clean'].str.replace('[^\w\s]', '') # for punctuation 
wine_reviews['description_clean'] = wine_reviews['description_clean'].str.lower() # for lower case 
```

```python
# Remove stopwords

from nltk.tokenize.treebank import TreebankWordDetokenizer
from nltk.corpus import stopwords
from nltk import word_tokenize          

stopwords = set(stopwords.words('english'))
# Detokenizer combines tokenized elements
detokenizer = TreebankWordDetokenizer()

# Define function to remove stopwords and change to lowercase and tokenize a text column
def clean_description(desc):
    desc = word_tokenize(desc.lower())
    desc = [token for token in desc if token not in stopwords and token.isalpha()]
    return detokenizer.detokenize(desc)
```

```python
# Use the function define above to create a clean column derived from the description column
wine_reviews["description_clean"] = wine_reviews["description_clean"].apply(clean_description)
```

### Distribution  of observations

As we saw from the data summary, there are numerous number of wines reviewed, some from different provinces, some from different wineries and so forth, but it is not necessary that the number of observations are equally distributed amony all differnt wine types. This can be a cause of concern later on when we build our model. Therefore, let us see firsthand how the observations are distributed.

```python
# Define function to categorise based on the number of observation available in the dataset
def observations(points):
    if points in range(0, 11):
        return '0 - 10'
    elif points in range(11, 101):
        return '11 - 100'
    elif points in range(101, 1001):
        return '101 - 1000'
    else:
        return '1000+'
```

```python
# Categorise each province, variety and winery based on the number of observations with the help of the function 

province['count_group'] = province['count'].apply(observations)
variety['count_group'] = variety['count'].apply(observations)
winery['count_group'] = winery['count'].apply(observations)

# Count number of province, variety and winery falling in each group
province_observations_frequency = province.groupby(['count_group']).count()
variety_observations_frequency = variety.groupby(['count_group']).count()
winery_observations_frequency = winery.groupby(['count_group']).count()
```

```python
# plot variety based on the number of observations in the dataset

order = ['1000+', '101 - 1000', '11 - 100', '0 - 10']
fig = plt.figure(figsize=(25, 6))

# Set seaborn style
sns.set(style="whitegrid")

# Province
plt.subplot(1, 3, 1)
ax = sns.barplot(x = province_observations_frequency['count'], y = province_observations_frequency.index, order = order )    
ax.set_title("Number of observations vs Number of Provinces")
plt.xlabel('Count of Provinces')
plt.ylabel('Number of Observations')

# Variety
plt.subplot(1, 3, 2)
ax = sns.barplot(x = variety_observations_frequency['count'], y = variety_observations_frequency.index, order = order )    
ax.set_title("Number of observations vs Number of Varieties")
plt.xlabel('Count of Varieties')
plt.ylabel('Number of Observations')

# Winery
plt.subplot(1, 3, 3)
ax = sns.barplot(x = winery_observations_frequency['count'], y = winery_observations_frequency.index, order = order )    
ax.set_title("Number of observations vs Number of Wineries")
plt.xlabel('Count of Wineries')
plt.ylabel('Number of Observations')

# plt.savefig('wine_observations.png', bbox_inches='tight')
plt.show()
```

# 3. Model Building

```python
# reference - https://stackoverflow.com/questions/47423854/sklearn-adding-lemmatizer-to-countvectorizer
from sklearn.feature_extraction.text import CountVectorizer
from nltk import word_tokenize          
from nltk.stem import WordNetLemmatizer 

# We will define a function for word lemmatization
class LemmaTokenizer(object):
    def __init__(self):
        self.wnl = WordNetLemmatizer()
    def __call__(self, articles):
        return [self.wnl.lemmatize(t) for t in word_tokenize(articles)]

# Define count vectorization for building final model
cv_new = CountVectorizer(tokenizer=LemmaTokenizer(),
                                strip_accents = 'unicode', # works 
                                stop_words = 'english', # works
                                lowercase = True, # works
                                max_df = 0.5, # works
                                min_df = 10) # works
```

```python
# Count vectorize the description and define training features
X_new = cv_new.fit_transform(wine_reviews['description_1'])
```

```python
# We have 8,337 features after processin our explanatory variable
X_new.shape
```

```python
# Error rate we use null model i.e. predict all wine varieties are from Pinot Noir (beacuse it has the maximum number of observations)
print(f"Error rate for our variety predicting null model = {1 - (13272/129971):.2f}")
```

We want to predict the **variety** of the wine and the **province** it was made in using the description of the wine given by the consumer/customer. Thus, we will be using the description of the wine given by the wine tasters to build our model.

```python
# Import relevant modules for modelling
from sklearn.model_selection import train_test_split, cross_validate, GridSearchCV
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.naive_bayes import MultinomialNB
```

## 3.1. Predict Variety 

```python
# Response variable is 'variety'
y_variety = wine_reviews['variety']

# 80-20% for train test split to predict points based on description
X_train_mnb_v2, X_test_mnb_v2, y_train_mnb_v2, y_test_mnb_v2 = train_test_split(X_new, y_variety, test_size=0.2, random_state=123)
```

### 3.1.1. Feature Selction

```python
# We have too many features, we'll try picking lesser features to build our model
from sklearn.feature_selection import SelectKBest, chi2

# Best 1000 feature
selector = SelectKBest(chi2, k=1000)
# help selector understand the data we are working with
selector.fit(X_train_mnb_v2, y_train_mnb_v2)
# We transform the data using variable, selector, that we defined earlie
X_train_mnb_v2_new = selector.transform(X_train_mnb_v2)
X_test_mnb_v2_new = selector.transform(X_test_mnb_v2)
```

### 3.1.2. Model selection & Hyperparameter tuning 

```python
# We chose 3 classifier models to do the model selection - Logistic Regression, Decision Tree and, Multinomial Naives Bayes 

models = {
    'LR': GridSearchCV(LogisticRegression(solver='lbfgs'),
                       param_grid = {'C': [0.01, 0.1, 1.0]},
                       cv=5#,verbose=10,n_jobs=-1
                        ),
    'DT': GridSearchCV(DecisionTreeClassifier(),
                       param_grid = {'max_depth': [30,40,50]},
                       cv=5),
    'NB': GridSearchCV(MultinomialNB(),
                       param_grid = {'alpha': [0.01, 0.1, 1, 10]},
                       cv=5#,verbose=10,n_jobs=-1
                      )}

print("*** Hyperparameter tuning ***")
for name, model in models.items():
    model.fit(X_train_mnb_v2_new, y_train_mnb_v2)
    print(f"{name} best hyperparams = {model.best_params_}.")
    print(f"{name} error: {1 - model.best_score_:.2f}")
                      
```

```python
# Our best model is the Logistic model with C = 1
model_mnb_v2 = LogisticRegression(solver='lbfgs',C=1.0)
```

```python
# Train model using Logistic model and calculate test error rate to check how good the model is
model_mnb_v2.fit(X_train_mnb_v2_new, y_train_mnb_v2)
cv_score_v2 = cross_validate(model_mnb_v2, X_train_mnb_v2_new, y_train_mnb_v2, cv=5)
print(f"CV error rate = {1 - cv_score_v2['test_score'].mean():.2f}")
print(f"Test Error rate = {1 - model_mnb_v2.score(X_test_mnb_v2_new, y_test_mnb_v2):.2f}")
```
